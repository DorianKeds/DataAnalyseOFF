# II - Data Cleaning

```python
df['Sugar_Value'] = df['Sugar_Value'].str.replace('g','')
df['Sugar_Value'] = df['Sugar_Value'].str.replace(',','.')
df['Sugar_Value'] = df['Sugar_Value'].str.replace('<','')
df['Sugar_Value'] = df['Sugar_Value'].str.replace('XXX','0')
df['Sugar_Value'] = df['Sugar_Value'].str.replace('~','')
df['Sugar_Value'] = df['Sugar_Value'].astype('float')

df['Energie(Kcal)_100'] = df['Energie(Kcal)_100'].str.replace(',','.')
df['Energie(Kcal)_100'] = df['Energie(Kcal)_100'].str.replace('kcal','')
df['Energie(Kcal)_100'] = df['Energie(Kcal)_100'].str.replace('XXX','0')
df['Energie(Kcal)_100'] = df['Energie(Kcal)_100'].astype('float')

df['Fat_Value'] = df['Fat_Value'].str.replace('g','')
df['Fat_Value'] = df['Fat_Value'].str.replace(',','.')
df['Fat_Value'] = df['Fat_Value'].str.replace('XXX','0')
df['Fat_Value'] = df['Fat_Value'].astype('float')

df['Saturated_Fatty_Acid_Value'] = df['Saturated_Fatty_Acid_Value'].str.replace('g','')
df['Saturated_Fatty_Acid_Value'] = df['Saturated_Fatty_Acid_Value'].str.replace(',','.')
df['Saturated_Fatty_Acid_Value'] = df['Saturated_Fatty_Acid_Value'].str.replace('XXX','0')
df['Saturated_Fatty_Acid_Value'] = df['Saturated_Fatty_Acid_Value'].astype('float')

df['Sodium_Value'] = df['Sodium_Value'].str.replace('g','')
df['Sodium_Value'] = df['Sodium_Value'].str.replace(',','.')
df['Sodium_Value'] = df['Sodium_Value'].str.replace('XXX','0')
df['Sodium_Value'] = df['Sodium_Value'].astype('float')

df['Quantity'] = df['Quantity'].str.replace(',','.')
df['Quantity'] = df['Quantity'].str.replace('ml','')
df['Quantity'] = df['Quantity'].str.replace('kg','')
df['Quantity'] = df['Quantity'].str.replace('g','')
df['Quantity'] = df['Quantity'].str.replace('l','')
df['Quantity'] = df['Quantity'].str.replace('-','')
df['Quantity'] = df['Quantity'].str.replace('75\xa0 ℮','')
df['Quantity'] = pd.to_numeric(df.Quantity, errors='coerce')
df.Quantity = df.Quantity.fillna(0)
```

# III - Data analysis

## 1. Distribution of the number of products by Nutri Score, Nova and Eco score categories ( 5*4*5 = 100 possibilities)

```python
groupby_eco = df.groupby(['Eco_Score'])['Product_Name'].count()
groupby_eco.plot(kind = 'bar')
```

```python
groupby_nutri = df.groupby(['Nutri_Score'])['Product_Name'].count()
groupby_nutri.plot(kind = 'bar', color = "blue")
```

```python
groupby_nova = df.groupby(['Nova_Score'])['Product_Name'].count()
groupby_nova.plot(kind = 'bar', color = "green")
```

## 2. Liste des produits ayant “gluten free” optimaux (Nutri score A, Nova 1 et Ecoscore A)

```python
nuti_a = df[df['Nutri_Score'] == 'A']
nova_1 = nuti_a[nuti_a['Nova_Score'] == '1 - Aliments non transformés ou transformés minimalement']
eco_A = nova_1[nova_1['Eco_Score'] == 'Eco-score A'].filter(items = ['Product_Name'])
eco_A
```

## 3. Nombre de produit dont le ratio sucre/produit est supérieur à 0.6

```python
df['Ratio_Su/Prod'] = df['Sugar_Value']/df['Quantity']
rat = df[df['Ratio_Su/Prod'] > 0.6]
rat.size
```

## 4. Liste des produits dont le nombre de calories pour 100g est supérieur à 500.

```python
cal_sup500 = df[df['Energie(Kcal)_100'] > 500].filter(items = ['Product_Name'])
cal_sup500
```

## 5. Liste des produits contenant de l’huile de palme, vendu en France et en Belgique

```python
oil_palm = df[df['Palm_Oil_Ingredient'].str.contains('Huile de palme')]
fr = oil_palm[oil_palm['Sales_countries'].str.contains('France')]
bl = fr[fr['Sales_countries'].str.contains('Belgique')]
products = bl.filter(items = ['Product_Name'])
products
```

##### Qu'est ce qu'un produit sain ? Un produit sain doit avoir un nutri score le plus faible ('A') et doit avoir une faible teneur en sucre, en sodium, en Graisse et en Acide gras

```python
nutriA = df[df.Nutri_Score == 'A'].nlargest(10, 'Fat_Value')
nutriA.filter(items = ['Product_Name', 'Fat_Value'])
```

```python
nutriA_Sat = df[df.Nutri_Score == 'A'].nlargest(10, 'Saturated_Fatty_Acid_Value')
nutriA_Sat.filter(items = ['Product_Name', 'Saturated_Fatty_Acid_Value'])
```

```python
nutriA_Su = df[df.Nutri_Score == 'A'].nlargest(10, 'Sugar_Value')
nutriA_Su.filter(items = ['Product_Name', 'Sugar_Value'])
```

```python
nutriA_So = df[df.Nutri_Score == 'A'].nlargest(10, 'Sodium_Value')
nutriA_So.filter(items = ['Product_Name', 'Sodium_Value'])
```
